```python
import pydicom
import matplotlib.pyplot as plt

dicom_file_path='data/test_dataset/LIDC-IDRI/LIDC-IDRI-0078/01-01-2000-NA-NA-68432/3166.000000-Segmentation of Nodule 1 - Annotation 4-05172/1-1.dcm'
dicom_data=pydicom.dcmread(dicom_file_path)

pixel_array=dicom_data.pixel_array
print(pixel_array.shape)

for i in range (len(pixel_array)):
    plt.imshow(pixel_array[i],cmap=plt.cm.gray)
    plt.title="DICOM Image"
    plt.axis('off')
    plt.show()
```
