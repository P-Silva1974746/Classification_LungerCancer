This is for only one image (one slice) not to a series of images

```python
import pydicom
import matplotlib.pyplot as plt

dicom_file_path='data/test_dataset/LIDC-IDRI/LIDC-IDRI-0001/01-01-2000-NA-NA-30178/3000566.000000-NA-03192/1-001.dcm'
dicom_data=pydicom.dcmread(dicom_file_path)

pixel_array=dicom_data.pixel_array


plt.imshow(pixel_array,cmap=plt.cm.gray)
plt.title="DICOM Image"
plt.axis('off')
plt.show()
```
