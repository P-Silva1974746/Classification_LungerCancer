```python
import SimpleITK as sitk
import matplotlib.pyplot as plt

#Read the series of images

reader= sitk.ImageSeriesReader()
dicom_names = reader.GetGDCMSeriesFileNames('data/test_dataset_2/LIDC-IDRI/LIDC-IDRI-0072/01-01-2000-NA-CT CHEST W CONT-45499/4.000000-Recon 3-88650')
reader.SetFileNames(dicom_names)
image =reader.Execute()

#convert the image to an array of slices then is equal as to it was before
image_array = sitk.GetArrayFromImage(image)

for slice_number in range(len(image_array)):
    plt.imshow(image_array[slice_number, :, :], cmap='gray')
    plt.title(f"DICOM Slice {slice_number}")
    plt.axis('off')
    plt.show()
```
